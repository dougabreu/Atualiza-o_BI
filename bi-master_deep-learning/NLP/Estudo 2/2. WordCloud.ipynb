```
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import urllib.request  
import bs4 as bs  
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```
raw_html = urllib.request.urlopen('https://blogs.oglobo.globo.com/miriam-leitao/post/azevedo-versus-trump-o-contexto.html')  
raw_html = raw_html.read()
article_html = bs.BeautifulSoup(raw_html, 'lxml')

article_paragraphs = article_html.article.find_all('p')
article_text = ''
for para in article_paragraphs:  
    article_text += para.text
```

```
stopwords = nltk.corpus.stopwords.words('portuguese')
wc = WordCloud(stopwords=stopwords, width=800, height=400)
wc.generate(article_text)
wc.to_file('output.png')
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
