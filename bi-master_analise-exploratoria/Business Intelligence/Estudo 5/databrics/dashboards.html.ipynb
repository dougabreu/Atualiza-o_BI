## Começando

Os painéis são compostos de elementos. Esses elementos são criados pela saída das células do notebook. Vamos criar alguns elementos (e células) para o painel que criaremos. A primeira célula cria um título de painel usando a função `` displayHTML () ``.

```
displayHTML("""<font size="6" color="red" face="sans-serif">Painel de análise de dados de compartilhamento de bicicletas""")
```

Crie um painel que exiba um conjunto de dados de compartilhamento de bicicletas disponível como um conjunto de dados hospedado pelo Databricks.

```
# Neste exemplo, usamos o conjunto de dados de compartilhamento de bicicletas disponível em `dbfs`
df = spark.read.format("csv").option("inferSchema", "true").option("header", "true").load("dbfs:/databricks-datasets/bikeSharing/data-001/day.csv")

df.registerTempTable("bikeshare")
```

```
df.head(100)
```

Crie um gráfico das condições da bicicleta em cada uma das estações. Crie também uma célula com esse rótulo para rotular corretamente o gráfico.

** Condições de bicicleta nas estações do ano **

```
display(spark.sql("SELECT season, MAX(temp) as temperature, MAX(hum) as humidity, MAX(windspeed) as windspeed FROM bikeshare GROUP BY season ORDER BY SEASON"))
```

## Criar um painel

Agora que temos alguns elementos para exibir, crie um painel a partir deles.

navegar para **View** menu e selecione **+ New Dashboard**.

![dashboard demo](https://docs.databricks.com/_static/images/dashboards/dashboard-demo-0.png)

Give your dashboard a name.

Por padrão, o novo painel inclui todas as células que você criou até o momento. Você pode reorganizar e remodelar cada célula como achar melhor. Navegue de volta para o código do notebook, selecionando a opção ** Código ** no menu ** Exibir **.

![something](https://docs.databricks.com/_static/images/dashboards/dashboard-demo-2.png)

Novas células não aparecem automaticamente no seu painel. Você os adiciona manualmente clicando no ícone do painel à direita e marcando a caixa de seleção ao lado do nome do painel.

![something](https://docs.databricks.com/_static/images/dashboards/dashboard-demo-3.png)

Você pode adicionar e remover células da mesma maneira.

## Adicionar gráficos a um painel

Crie mais gráficos para adicionar ao painel. Primeiro, adicione um gráfico das condições de ciclismo em todos os meses em nosso conjunto de dados. Crie também uma célula com um rótulo. Adicione as células ao painel.

**Condições médias de bicicleta**

```
display(spark.sql("SELECT mnth as month, AVG(temp) as temperature, AVG(hum) as humidity, AVG(windspeed) as windspeed FROM bikeshare GROUP BY month ORDER BY month"))
```

Embora a média seja agradável de ver, é provável que estejamos mais preocupados com as condições extremas. Crie um gráfico das condições extremas da bicicleta ao longo dos meses. Crie também uma célula com um rótulo. Adicione as células ao painel.

**Extreme Biking Conditions**

```
%sql SELECT mnth as month, MAX(temp) as max_temperature, MAX(hum) as max_humidity, MAX(windspeed) as max_windspeed FROM bikeshare GROUP BY mnth ORDER BY mnth
```

Agora que você criou um painel, pode organizá-lo. Navegue para a parte superior do novo painel. Selecione os cantos inferiores esquerdo ou direito para redimensionar e remodelar cada bloco no painel. As células de marcação decrescente se tornam rótulos para cada seção do painel..
